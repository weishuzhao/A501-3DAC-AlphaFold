```python
from pymol import cmd
import sys
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def percent_aa(ss_string,aatype):
    # H percent, S percent, L percent
    return ss_string.count(aatype)/len(ss_string)

aadic = "ACDEFGHIKLMNPQRSTVWY"
```

```python
avaiable_DF = pd.read_csv("available_result.csv")
```

```python
list_aa_A501 = [[] for i in range(20)]
list_aa_3DAC = [[] for i in range(20)]
for i in range(20):
    for j in range(avaiable_DF.shape[0]):
        list_aa_A501[i].append(percent_aa(avaiable_DF["A501_sequence"][j],aadic[i]))
        list_aa_3DAC[i].append(percent_aa(avaiable_DF["3DAC_sequence"][j],aadic[i]))
```

```python
plt.figure(figsize=(12,8))
for i in range(20):
    plt.subplot(4,5,i+1)
    sns.kdeplot(list_aa_A501[i],shade=True,label="A501")
    sns.kdeplot(list_aa_3DAC[i],shade=True,label="3DAC")
    plt.text(0.2,40,aadic[i],fontsize=14)
    plt.xlim(0,0.3)
    plt.ylim(0,60)
    plt.ylabel("")
plt.legend(loc='lower right')
plt.tight_layout(pad=0.5)
plt.savefig("Figure/aa_count.png",dpi=300)
```
