```python
from pymol import cmd
import sys
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

```

```python
f = open("RMSD/RMSD_A501_non_001.txt")
total_list = []
line = f.readline()
while line:
    line = line.replace("(",",")
    line = line.replace(")\n","")
    line_list = line.split(",")
    if line_list[2] != "No Data":
        total_list.append(line_list)
    line = f.readline()
```

```python
plt.figure(figsize=(20,4))
plt.plot([item[8] for item in total_list])
plt.plot([item[3] for item in total_list])
```
